```python
%%R

# My commonly used R imports

require('tidyverse')
library(ggbump)
library(ggbeeswarm)
library(lubridate)
library(dplyr)
library(tidyr)
```

```python
%%R
 
# Import data with R
salt <- read_csv('./data/DSNY_Salt_Usage.csv', show_col_types = FALSE)
salt$`Date of Report` <- as.Date(salt$`Date of Report`, format = "%m/%d/%Y")
salt %>% head(4)
```

```python
%%R

# calculate the percentage of salt used by each borough
library(dplyr)

salt_pct <- salt %>%
  mutate(year = case_when(month(`Date of Report`) >= 6 ~ year(`Date of Report`), 
                          TRUE ~ year(`Date of Report`) - 1)) %>%
  group_by(year) %>%
  summarise_at(vars(Manhattan:`Staten Island`), list(~ sum(.)/sum(`Total Tons`))) %>%
  pivot_longer(cols = Manhattan:`Staten Island`, names_to = "borough", values_to = "percentage")

salt_pct
```

```python
%%R

ggplot(salt_pct, aes(x = year, y = percentage, fill = borough)) +
  geom_bar(stat = "identity", position = "fill") +
  scale_fill_brewer(palette = "Dark2") +
  xlab("Year") +
  ylab("Percentage") +
  ggtitle("Queens used most road salt in each season") +
  theme(plot.title = element_text(hjust = 0.5),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        axis.line = element_line(colour = "black"),
        text = element_text(size=12)) +
  coord_flip() +
  scale_x_reverse()
```

### A line chart exclusively to 2021

```python
%%R
# Filter data by date range
salt_filtered <- salt %>%
  filter(`Date of Report` >= as.Date("2021-11-30") & `Date of Report` <= as.Date("2022-03-31"))

# Create line chart
ggplot(salt_filtered, aes(x = `Date of Report`, y = `Total Tons`)) +
  geom_line() +
  scale_color_manual(values = c("blue", "red", "green", "orange", "purple"),
                     labels = c("Manhattan", "Bronx", "Brooklyn", "Queens", "Staten Island")) +
  labs(x = "Date of Report", y = "Total Tons", title = "Total Tons by Borough") +
  theme_minimal()
```
